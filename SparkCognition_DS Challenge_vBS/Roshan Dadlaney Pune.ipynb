Ans: We can determing the number of clusters to be used using elbow method

Then plot a graph of Sum of squared distances(Sum of distances of all points from the centroid within a cluster) vs K

Metrics of Kmeans to evaluate it:
1)Inertia of Sum of squared errors: It calculates the sum of distances of all the points within a cluster from the centroid of that cluster. It should be as low as possible
2)Dunn Index: It is the ratios of the min(inter cluster distance)/max(intra cluster distance)
We want to maximize Dunn index

Exercise 4: Anomaly Detection

```python
import pandas as pd
import numpy as np
```

```python
ana=pd.read_csv('anomaly_detection.csv')
```

Taking the last 91 days(1st Oct to 30th Dec) as test data and the days before as train data

```python
ana_train=ana[:274]
ana_test=ana[274:]
```

```python
ana_test.describe()
```

Applying PCA and reducing the dimensions to 3

```python
from sklearn.decomposition import PCA

pca_ana = PCA(n_components=3, svd_solver='full', random_state=2021)
pca_ana_train = pca_ana.fit_transform(ana_train)
```

```python
pca_ana_train=pd.DataFrame(pca_ana_train)
pca_ana_train.index=ana_train.index
```

```python
pca_ana_train.head()
```

```python
def ana_risk(orig_df,reduced_df):
    risk = np.sum((np.array(orig_df)-np.array(reduced_df))**2, axis=1)
    risk = pd.Series(data=risk,index=orig_df.index)
    risk = (risk-np.min(risk))/(np.max(risk)-np.min(risk))
    return risk
```

```python
pca_ana_train_inverse = pca_ana.inverse_transform(pca_ana_train)
pca_ana_train_inverse = pd.DataFrame(data=pca_ana_train_inverse,index=ana_train.index)
```

```python
import matplotlib.pyplot as plt
ana_score=ana_risk(ana_train,pca_ana_train_inverse)
#print(ana_score)
#plt.plot(ana_score)
#ana_score[ana_score>0.3]
ana_score[40:52]
```
